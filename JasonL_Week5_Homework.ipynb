12. Unpack the array tuples from the above 'reshapedArray'  into 4 well named variables. Print the 4 variables.

````python
# Question 12
import numpy as np

arrayA = ([[25, 16]])
arrayB = ([[11, 2], [13, 4]])
arrayC = ([[7, 81], [5, 6], [11, 12]])

# Bring the previous mess of arrays together
newArray = np.concatenate((arrayA, arrayB, arrayC))
# reshape the new array into 4 tuples
reshapedArray = newArray.reshape(4, -1)

# unpack array into 4 variables
tuple0, tuple1, tuple2, tuple3 = reshapedArray
print("Tuple0: ", tuple0)
print("Tuple1: ", tuple1)
print("Tuple2: ", tuple2)
print("Tuple3: ", tuple3)
````

14. Take 'comboArray' and perform the following slicing activities:
    - print sec1 - the 2nd element
    - print sec2 - all elements from the 3rd element to the last
    - print sec3 - all elements from the 4th to the 14th elements
    - print sec4 - the last 6 elements
    - print sec5 - all element from #0 up to and including #15, using the negative number method, i.e. taking a section from the end.
    - print sec6 - from #20 every even element to the end
    - print sec7 - from the last element moving forward, every 5th element.

````python
# Question 14
import numpy as np

one = ([10, 11, 12, 13, 14, 15, 16, 17])
two = ([20, 21, 22, 23, 24, 25, 26, 27])
three = ([ 0, 1, 2, 3, 4, 5, 6, 7])

# combine arrays
comboArray = np.concatenate((one, two, three))

# sort array
comboArray = np.sort(comboArray, axis= -1)

print("\nSorted and concatenated array:\n",comboArray, "\n")
print("sec1: ",comboArray[1:2])
print("sec2: ",comboArray[2:])
print("sec3: ",comboArray[3:14])
print("sec4: ",comboArray[-6:])
print("sec5: ",comboArray[-27:15])
print("sec6: ",comboArray[20::2])
print("sec7: ",comboArray[27::-5])
````

15. Using `Series`, create a `DataFrame` that looks like this:

    | Ingredients | Quantity | Unit |
    |----|----|----|
    | Flour | 4 | cups |
    | Milk | 1 | cup |
    | Eggs | 2 | large |
    | Spam | 1 | can |

    Name: Dinner, dtype: object

````python
# Question 15
import pandas as pd

ingredients = pd.Series(["Flour", "Milk", "Eggs", "Spam"], name = 'Ingredients')
quantity = pd.Series([4, 1, 2, 1], name = 'Quantity')
unit = pd.Series(["cups", "cup", "large", "can"], name = "Unit")

dinner = pd.concat([ingredients, quantity, unit], axis = 1)

print(dinner)
````
