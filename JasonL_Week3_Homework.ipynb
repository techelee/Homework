```python
# Question 10
newList = ["string", 3.14, 42, True, ["apple", "banana", "cherry"]]
print("newList: ",newList)
newList[2] = 53
print("Modified:",newList)
newList.append("extra")
newList[2:1] = ["inserted"]
print("After insert and append:",newList)
print("newList length:", len(newList))

# create a subset of newList with onlty string types
newListStrings = [newList[0], newList[2], newList[5], newList[6]]
print("\nnewListStrings:", newListStrings)

# Extend the list
newList = newList + newListStrings
print("\nExtended list:", newList)

# new "simList" then sort it
simList = ["Gek", "Vykeen", "Korvax", "Autophage", "Sentinel", "Anomaly"]
simList.sort()
print("\nSorted simList:", simList)

# combine second list with sorted simList
simList2 = simList.copy()
newSimList = newListStrings + simList2
print("\nCombining newListStrings and SimList2:", newSimList)
```

```python
# Question 11
myTuple = ("Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune")
tripleTuple = 3 * myTuple
print("tripleTuple:", tripleTuple)
print("12th element of tripleTuple:", tripleTuple[11])

# tuples are immutable, so tuple sort is not permanent
print("\ntripleTuple sorted:", sorted(tripleTuple))

thirdTuple = tripleTuple[3], tripleTuple[5], tripleTuple[7], tripleTuple[9]

# Unpack thirdTuple into variables
planet1, planet2, planet3, planet4 = thirdTuple
print("\nUnpacked planets from thirdTuple:", planet1, planet2, planet3, planet4)

#fourth tuple with a single element
fourthtuple = 50
print("\nfourthtuple value:", fourthtuple)

# add 2nd and 3rd tuple together into fifth tuple and print it
fifthTuple = tripleTuple + thirdTuple
print("\nfifthTuple:", fifthTuple)
```

```python
# Question 12
firstSet = {"monkey", "fruit bat", "Malayan tapir"}
fruitList = ["apple", "banana", "cherry"]
firstSet.update(fruitList)
print("After adding fruits:", firstSet)

firstSet.add("Mustang")
print("After adding car:", firstSet)

secondSet = {"blue", "washer", "chair"}
thirdSet = firstSet.union(secondSet)
print("\nUnion of firstSet and secondSet:", thirdSet)

secondSet.pop()
print("\nAfter popping an element from secondSet:", secondSet)

firstSet.clear()
print("\nAfter clearing firstSet:", firstSet)

thirdSet.discard("Malayan tapir")
thirdSet.remove("blue")
print("\nAfter removing elements from thirdSet:", thirdSet)
```

```python
# Question 13
newDictionary = {
    "name": "CSI",
    "street number": 2700,
    "street": True,
    "city, state": ["Springfield", "MO"],
    "zip": 65803.4130
}
print("newDictionary:", newDictionary)
print("\ncity, state:", newDictionary["city, state"])

newDictionary["name"] = "Jason"
print("\nnewDictionary with the name changed:", newDictionary)

newDictionary["color"] = "green"
print("\nnewDictionary with favorite color:", newDictionary)

#Add a list, tuple or set to your dictionary
newDictionary["list"] = (1, 2, 3)
print("\nnewDictionary with list:", newDictionary)

#Print a list of the dictionary keys
print("\nnewDictionary keys:", newDictionary.keys())

#Print a list of the dictionary values
print("\nnewDictionary values:", newDictionary.values())

#Copy your 1st dictionary into a 2nd dictionary
secondDictionary = newDictionary.copy()

#Pop an item from the 2nd dictionary, and print the dictionary
secondDictionary.pop("street")
print("\nsecondDictionary without street:", secondDictionary)

#Remove all the elements from the 2nd dictionary and print the result
secondDictionary.clear()
print("\nsecondDictionary cleared:", secondDictionary)
```
